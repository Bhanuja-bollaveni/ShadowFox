WHAT IS MATPLOTLIB? 
Matplotlib is a comprehensive, open-source Python library for creating static, animated, and interactive visualizations. It is a foundational tool in the Python data science ecosystem, designed to produce publication-quality figures. 
Key Features and Uses
Versatile Plotting: Matplotlib can generate a wide range of plot types with just a few lines of code, including line plots, scatter plots, bar charts, histograms, pie charts, and 3D plots.
High Customizability: The library offers extensive control over virtually every aspect of a plot, from line styles and colors to font properties, axes, and legends, allowing users to tailor visualizations precisely.
Integration: It integrates seamlessly with the Python scientific computing stack, most notably with NumPy for numerical operations and Pandas for data manipulation.
Multiple Interfaces: Matplotlib offers two main APIs:
pyplot (MATLAB-style interface): A procedural, state-machine interface that is easy for beginners and for quick, interactive plotting.
Object-Oriented (OO) interface: Provides more explicit control by working directly with Figure and Axes objects, which is recommended for more complex plots or embedding plots into applications.
Cross-Platform Compatibility: It can be used across various environments, including Python scripts, IPython shells, Jupyter Notebooks, and web application servers.
GRAPH TYPES 
1.LINE PLOT
A line plot in Matplotlib is a graph that connects a series of data points with a line to display a sequence of values, which helps in identifying trends and patterns in data over a continuous range, such as time. The primary function used is matplotlib.pyplot.plot(). 
Purpose: Line plots are ideal for visualizing continuous data, time-series information, or mathematical functions.
Mechanism: The plt.plot() function takes x and y values (typically as NumPy arrays or lists) and draws lines between consecutive points.
Customization: You can customize the appearance using various parameters, including color, linewidth, linestyle, and marker to add points at the data locations.
EXAMPLE:
In this example, a simple line chart is generated using NumPy to define data values. The x-values are evenly spaced points, and the y-values are calculated as twice the corresponding x-values.
  

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4])   
y = x*2  
plt.plot(x, y)  
plt.show()
```

This is a basic line plot where x contains four points and y is calculated as twice of each x value.plt.plot() creates the line and plt.show() renders the plot.
2.BAR CHART
In this there are 2 types of bars they are pytlot.bar() function for vertical bars and pyplot.barh() for horizontal bars
A bar chart is a graphical representation that uses rectangular bars to compare different categories. The height or length of each bar corresponds to the value it represents. The x-axis typically shows the categories being compared, while the y-axis shows the values associated with those categories. This visual format makes it easy to compare quantities across different groups.While Seaborn itself does not have a dedicated pie chart function, you can easily create one by using the matplotlib.pyplot.pie() function and leveraging Seaborn's attractive color palettes.
Bar plots are significant because they provide a clear and intuitive way to visualize categorical data. They allow viewers to quickly grasp differences in size or quantity among categories, making them ideal for presenting survey results, sales data, or any discrete variable comparisons.
EXAMPLE:
A simple example where we visualize the marks of a bhanuja in different years

```python
import matplotlib.pyplot as plt
import numpy as np
Years = ['2020', '2021', '2022', '2023']
Marks = [400, 350, 300, 450]
plt.bar(Years, Marks, color='Pink')
plt.title('Bhanuja Marks')
plt.xlabel('Years')
plt.ylabel('Marks')
plt.show()
```

3.PIE CHART 
A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data.  Pie charts in Python are widely used in business presentations, reports, and dashboards due to their simplicity and effectiveness in displaying data distributions
A pie chart consists of slices that represent different categories. The size of each slice is proportional to the quantity it represents. The following components are essential when creating a pie chart in Matplotlib:
Data: The values or counts for each category.
Labels: The names of each category, which will be displayed alongside the slices.
Colors: Optional, but colors can be used to differentiate between slices effectively.
Pie charts provide a visual representation of data that makes it easy to compare parts of a whole. They are particularly useful when:
Displaying relative proportions or percentages.
Summarizing categorical data.
Highlighting significant differences between categories.
EXAMPLE:

```python

from matplotlib import pyplot as plt
import numpy as np
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
fig = plt.figure(figsize=(10, 7))
plt.pie(data, labels=cars)
plt.show()
```

4.SCATTER PLOTS 
Scatter plots are one of the most fundamental tools for visualizing relationships between two numerical variables. matplotlib.pyplot.scatter() plots points on a Cartesian plane defined by X and Y coordinates. Each point represents a data observation, allowing us to visually analyze how two variables correlate, cluster or distribute.
EXAMPLE:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([16, 43, 6, 33, 69, 52, 76, 5, 25, 89])
y = np.array([79, 31, 74, 53, 20, 88, 43, 60, 32, 70])

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

plt.scatter(x,y) creates a scatter plot on 2D plane to visualize the relationship between 2 variables,with a title and axis labels added for clarity and context
5.HISTOGRAM 
Histograms are one of the most fundamental tools in data visualization. They provide a graphical representation of data distribution, showing how frequently each value or range of values occurs. Histograms are especially useful for analyzing continuous numerical data, such as measurements, sensor readings, or experimental results.
A histogram is a type of bar plot where:
The X-axis represents intervals (called bins) of the data.
The Y-axis represents the frequency of values within each bin.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=50, color='VIOLET', edgecolor='BLACK')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

WHAT IS SEABORN?
Seaborn is a popular Python library for creating informative and attractive statistical graphics. Built on top of the foundational Matplotlib library and tightly integrated with Pandas data structures, it provides a high-level interface that simplifies complex data visualization tasks. 
Key Features and Purpose
The primary goal of Seaborn is to make visualization a central part of exploring and understanding data by offering a dataset-oriented API and opinionated defaults. 
High-Level Interface: Seaborn's functions are designed to create complete, aesthetically pleasing plots with minimal code, abstracting away much of the complexity of Matplotlib.
Statistical Plotting: It specializes in visualizing relationships and distributions, automatically performing statistical estimation (like calculating confidence intervals and fitting regression models).
Integration with Pandas and NumPy: Seaborn works seamlessly with Pandas DataFrames and NumPy arrays, allowing users to quickly generate plots directly from their data analysis structures.
Aesthetics and Themes: The library offers a variety of built-in themes and color palettes (e.g., darkgrid, whitegrid, dark, white, ticks) to enhance the visual appeal of plots without manual customization.
Faceting and Multi-Plot Grids: It supports "faceting," which creates multiple plots based on different subsets or categories within a dataset, a powerful feature for comparing data across various conditions.
GRAPH TYPES:
1.LINE PLOT
To create a line plot in Python using the Seaborn library, you primarily use the seaborn.lineplot() function. This function is designed to show the relationship between two variables, especially over a continuous range like time.


```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("iris")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
```

2.BAR CHART
Seaborn provides the sns.barplot() function to create a bar chart, which is primarily used to display an estimate of central tendency (like the mean) for a numeric variable across different categories of a categorical variable. A barplot is basically used to aggregate the categorical data according to some methods and by default it's the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')
sns.barplot(x = 'who',
            y = 'fare',
            data = df)
plt.show()
```

3.PIE CHART
While Seaborn itself does not have a dedicated pie chart function, you can easily create one by using the matplotlib.pyplot.pie() function and leveraging Seaborn's attractive color palettes.


```python
import matplotlib.pyplot as plt
import seaborn as sns
data = [44, 45, 40, 41, 39]
labels = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5']
colors = sns.color_palette('bright')
plt.pie(data, labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```

4.SCATTER PLOT
To create a scatter plot in Seaborn, you primarily use the seaborn.scatterplot() or seaborn.relplot() functions. These functions help visualize the relationship between two numerical variables in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
sns.scatterplot(x="timepoint", y="signal", data=fmri)
plt.title("FMRI Signal Over Time")
plt.show()
```

5.HISTOGRAM
In Seaborn, you can create a histogram using the sns.histplot() function, which is the primary tool for visualizing the distribution of one or more numerical variables.

```python
import seaborn as sns
import numpy as np
import pandas as pd
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var, name = "Numerical Variable")
sns.histplot(data = num_var, kde = True)
```

MATPLOTLIB VS SEABORN 
Matplotlib: It is a Python library used for plotting graphs with the help of other libraries like Numpy and Pandas. It is a powerful tool for visualizing data in Python. It is used for creating statistical inferences and plotting 2D graphs of arrays. It was first introduced by John D. Hunter in 2002. It uses Pyplot to provide a MATLAB-like interface free and open-source. It is capable of dealing with various operating systems and their graphical backends.
Seaborn: It is also a Python library used for plotting graphs with the help of Matplotlib, Pandas, and Numpy. It is built on the roof of Matplotlib and is considered as a superset of the Matplotlib library. It helps in visualizing univariate and bivariate data. It uses beautiful themes for decorating Matplotlib graphics. It acts as an important tool in picturing Linear Regression Models. It serves in making graphs of statical Time-Series data. It eliminates the overlapping of graphs and also aids in their beautification.
1. Ease of Use
Matplotlib requires more lines of code to create graphs and customize them.
It is flexible but can be slightly difficult for beginners.
Seaborn is built on top of Matplotlib and provides simpler syntax.
It allows users to create attractive graphs with less code.
Conclusion:
Seaborn is easier to use than Matplotlib, especially for beginners.
2. Customization
Matplotlib offers full control over every element of a plot, including colors, labels, ticks, and layouts.
It is highly customizable but requires manual effort.
Seaborn provides good default themes and styles, but fine-grained control is limited compared to Matplotlib.
Conclusion:
Matplotlib offers better customization, while Seaborn focuses on simplicity and aesthetics.
3. Interactivity
Matplotlib produces mostly static plots with limited interactivity.
Interactive features require additional libraries or configurations.
Seaborn also produces static plots and relies on Matplotlib for rendering.
It does not support advanced interactivity by default.
Conclusion:
Both Matplotlib and Seaborn have limited interactivity.
4. Performance with Large Datasets
Matplotlib performs well with large datasets but may become slow when rendering complex plots.
Seaborn works efficiently with medium-sized datasets and is optimized for statistical visualization.
For extremely large datasets, performance can reduce in both libraries.
Conclusion:
Both libraries handle large datasets reasonably well, but Matplotlib offers more control for performance tuning.
5. Use Cases
Matplotlib is best suited for:
Custom visualizations
Low-level plotting
Scientific and engineering plots
Seaborn is best suited for:
Statistical analysis
Data exploration
Quick and attractive visualizations
