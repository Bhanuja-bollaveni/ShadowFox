```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\bhanu\Downloads\Product-Sales-Region.xlsx")
df.columns = df.columns.str.strip()

sales_col = df.select_dtypes(include="number").columns[0]

df.groupby("Product")[sales_col].sum().plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel(sales_col)
plt.show()
```

This bar chart shows the total sales generated by each product.
It helps identify the highest and lowest selling products.


```python
df.groupby("Product")[sales_col].sum().plot(kind="line", marker="o")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel(sales_col)
plt.xticks(rotation=45)
plt.show()
```

THe line chart represents sales trends over time.It helps in understanding whether sales are increasing or decreasing

```python
plt.hist(df[sales_col], bins=10)

plt.title("Sales Distribution")
plt.xlabel(sales_col)
plt.ylabel("Frequency")
plt.show()
```

This histogram shows how profit values are distributed.Most profit values fall within a specific range

```python
df.groupby("Product")[sales_col].sum().plot(kind="pie", autopct="%1.1f%%")

plt.title("Sales Share by Product")
plt.ylabel("")  
plt.show()
```

This pie chart represents the contribution of each region to total sales.It helps understand regional performance.
